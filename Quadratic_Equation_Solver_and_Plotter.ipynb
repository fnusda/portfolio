```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

valueA = float(input("Please enter the value of a: "))
valueB = float(input("Please enter the value of b: "))
valueC = float(input("Please enter the value of c: "))

discriminant = (valueB ** 2) - (4 * valueA * valueC)

if discriminant > 0:
    x1 = (-valueB + sqrt(discriminant)) / (2 * valueA)
    x2 = (-valueB - sqrt(discriminant)) / (2 * valueA)
    print(f"Two distinct real roots: x1 = {x1:.4f}, x2 = {x2:.4f}")
elif discriminant == 0:
    x1 = -valueB / (2 * valueA)
    print(f"One repeated root: x = {x1:.4f}")
else:
    print(f"No real roots (discriminant = {discriminant:.4f})")

# --- Graph ---
vertex = -valueB / (2 * valueA)
x = np.linspace(vertex - 6, vertex + 6, 400)   # 400 points around the vertex
y = valueA * x**2 + valueB * x + valueC

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='steelblue', linewidth=2, label=f'{valueA}x² + {valueB}x + {valueC}')
plt.axhline(0, color='black', linewidth=0.8)    # x-axis
plt.axvline(0, color='black', linewidth=0.8)    # y-axis
plt.grid(True, linestyle='--', alpha=0.5)

# Plot roots as red dots if they exist
if discriminant > 0:
    plt.scatter([x1, x2], [0, 0], color='crimson', zorder=5, label=f'Roots: {x1:.2f}, {x2:.2f}')
elif discriminant == 0:
    plt.scatter([x1], [0], color='crimson', zorder=5, label=f'Root: {x1:.2f}')

plt.title("Quadratic Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.tight_layout()
plt.show()
```
