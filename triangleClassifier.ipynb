```python
while True:
    angleAlpha = float(input("Enter the value of alpha angle: "))
    angleBeta  = float(input("Enter the value of beta angle: "))
    angleGama  = float(input("Enter the value of gama angle: "))

    if (angleAlpha + angleBeta + angleGama) != 180 or any(a <= 0 for a in [angleAlpha, angleBeta, angleGama]):
        print("Enter valid values\n")
        continue  # goes back to the top of the loop

    if angleAlpha == 90 or angleBeta == 90 or angleGama == 90:
        print("Right Triangle")
    elif angleAlpha > 90 or angleBeta > 90 or angleGama > 90:
        print("Obtuse Triangle")
    else:
        print("Acute Triangle")

    break  # valid input was processed, exit the loop
```
